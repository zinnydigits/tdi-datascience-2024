```python
# import json to read file
import json
```

```python
# load data from json file
with open('employees_data.json', 'r') as file:
    data = json.load(file)
# accessing data
for item in data:
    for key, value in item.items():
        print(f"{key}: {value}")
    print()
```

#### 1. Determine the total number of unique departments within the company.

```python
# use set to get unique departments in the dataset
dept = {data["department"] for data in data}
print("These are the departments in the company: ")
i = 1
for item in dept:
    print(f"{i}. {item}")
    i = i + 1
print(f"Therefore, there are {i-1} departments in the company.")
```

#### 2. List the top 5 highest-paid employees, sorted in descending order by their monthly salary

```python
# use lamda function to sort the employee's salary
sorted_employee = sorted(data, key=lambda x: x['salary'], reverse= True)

# use slicing to get the top 5 salaries
sorted_employee = sorted_employee[:5]
print("Top 5 Highest Paid Employees: ")
i = 1

# iterate to print employee's name and salary
for employee in sorted_employee:
    print(f"{i}. {employee['fullName']}: {employee['salary']}")
    i += 1
```

#### 3. Compute and present the average monthly salary of all employees in the company.

```python
# create an empty list
salaries = []

for item in data:
    # slice the salary to remove naira symbol and convert datatype to integer
    data_salary = int(item["salary"][1:])
    # add salary to salaries list
    salaries.append(data_salary)

# calculate the average salary
avg_salary = sum(salaries)/len(salaries)

print(f"The average salary of all employees in the company is N{avg_salary}")
    
```

#### 4. Show the number of employees in each department.

```python
grouped_data = {}
for item in data:
    department = item["department"]
    if department in grouped_data:
        grouped_data[department] += 1
    else:
        grouped_data[department] = 1
print("Department: Total No of Employees\n")
for key, value in grouped_data.items():
    print(f"{key}: {value}")
```

#### 5. Provide a list of employees ranked by their salary within each department, from highest to lowest.

```python
employees_sorted = sorted(data, key=lambda x: (x["department"], x["salary"]), reverse=True)

# group employees by department
grouped_employees = {}
for employee in employees_sorted:
    dept = employee["department"]
    if dept not in grouped_employees:
        grouped_employees[dept] = []
    grouped_employees[dept].append(employee)

# print sorted employees for each department
for dept, emp_list in grouped_employees.items():
    print(f"Department: {dept} \n")
    for emp in emp_list:
        print(f"{emp['fullName']} Salary: {emp['salary']}")
    print()
```

#### 6. Calculate the total monthly payroll expense for the company.

```python
# create an empty list
salaries = []

for item in data:
    # slice the salary to remove naira symbol and convert to integer
    data_salary = int(item["salary"][1:])
    # add salary to salaries list
    salaries.append(data_salary)

# calculate the average salary
total_salary = sum(salaries)

print(f"The total salary of all employees in the company is: N {total_salary}.")
    
```

#### 7. Determine the highest and lowest salary in the company and identify the corresponding employees.

```python
# use lamda function to sort the employee's salary
sorted_employee = sorted(data, key=lambda x: x['salary'])

# use slicing to get last 5 salaries since it was sorted in ascending order
highest_paid = sorted_employee[-1]
print("Highest Paid Employee:")
print(f"{highest_paid["fullName"]},  {highest_paid["salary"]}\n")

# use slicing to get the least paid employee
least_paid = sorted_employee[0]
print("Least Paid Employee:")
print(f"{least_paid["fullName"]},  {least_paid["salary"]}")
```

#### 8. Remove all employees with a Junior experience level and calculate the number of remaining employees.

```python
# set counter equal zero
i = 0

# iterate over dataset
for item in data:
    # conditional statement
    if item["experienceLevel"]  != "Junior":
        # count if condition was met
        i = i + 1
print(f"The total number of employees excluding the juniors is: {i}.")
```
